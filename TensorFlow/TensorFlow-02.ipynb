```python
dataFrame=dataFrame.drop("transmission",axis=1)
```

```python
dataFrame
```

```python
y=dataFrame["price"].values
```

```python
x=dataFrame.drop("price",axis=1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
#batch size aynı anda kaç veri verileceğini seçer validation data doğrulama datası
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),epochs=300,batch_size=250) 
```

```python
kayipverisi=pd.DataFrame(model.history.history)
```

```python
kayipverisi
```

```python
kayipverisi.plot()#val loss ayrılırsa overfitting yaptığın anlama gelir şuan tablo güzel
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi)
```
